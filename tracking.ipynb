```python
import os
from os import path
import shutil
import random

random.seed(0)

DATA_PATH = "E:\\GENA_RT\\DATA\\agro_hack"
TARGET_PATH = "E:\\GENA_RT\\DATA\\agro_hack\\video\\frames"

RATIO = 0.2

chunks = [
    '20220719_183951',
    '20220720_104238',
    '20220724_154038',
    '20220724_171831',
]

chunk = chunks[3]
chunk_path = path.join(DATA_PATH, chunk, chunk)

folders = os.listdir(chunk_path)
folders = list(map(path.basename, folders))

folders = list(sorted(map(lambda s: int(s[s.index('_')+1:]), folders)))
len(folders)
```

```python
to_vid = folders[5000:5500]
to_vid = [f'package_{i}' for i in to_vid]
#to_vid
```

```python
if os.path.exists(TARGET_PATH) and os.path.isdir(TARGET_PATH):
    shutil.rmtree(TARGET_PATH)

if not path.exists(TARGET_PATH):
    os.makedirs(TARGET_PATH)
```

```python
for i, frame in enumerate(to_vid):
    name = f"{chunk}-{frame}.png"
    shutil.copyfile(
        path.join(chunk_path, frame, 'color_frame.png'),
        #path.join(TARGET_PATH, f'{str(i).zfill(6)}.png')
        path.join(TARGET_PATH, name.replace('.png', '_png.png'))
    )
```
